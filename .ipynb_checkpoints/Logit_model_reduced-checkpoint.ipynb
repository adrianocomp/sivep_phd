```python
from imblearn.combine import SMOTEENN
from sklearn.model_selection import train_test_split
y2=df2['death']
X2=df2.drop(['death'],axis=1)

X_train2,X_test2,y_train2,y_test2=train_test_split(X2,y2,test_size=0.3,random_state=42)
smt = SMOTEENN()

# from sklearn.preprocessing import StandardScaler

# #scaler = StandardScaler()
# #X_train2 = scaler.fit_transform(X_train2)
# #X_test2 = scaler.transform(X_test2)

X_train_smote2, y_train_smote2 = smt.fit_resample(X_train2,y_train2)

from sklearn.linear_model import LogisticRegression
from sklearn import metrics
logit_model2 = LogisticRegression()
logit_model2.fit(X_train_smote2,y_train_smote2)

from sklearn.model_selection import cross_val_score
resultado_cv = cross_val_score(logit_model2, X_train_smote2, y_train_smote2, cv=10)

resultado_cv
```

```python
prediction2=logit_model2.predict(X_test2)
print('The accuracy of the Logistic Regression is',metrics.accuracy_score(prediction2,y_test2))
```

```python
coefficients2 = logit_model2.coef_[0]

feature_importance2 = pd.DataFrame({'Feature': X2.columns, 'Importance': np.abs(coefficients2)})
feature_importance2 = feature_importance2.sort_values('Importance', ascending=True)
feature_importance2.plot(x='Feature', y='Importance', kind='barh', figsize=(15, 15))
```
