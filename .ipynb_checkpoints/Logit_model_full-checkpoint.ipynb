```python
import pandas as pd
import numpy as np
df = pd.read_spss("bases_de_dados/Sivep/sivep_kids.sav")
```

```python
df.head()
```

```python
df = df.drop(['source_id_SIVEP','BD',
 'outcome_final_SIVEP',
 'outcome_geral_severe',
 'death2','UTI',
 'suport_vent',
 'Tempo_desfecho0',
 'SEM_NOT', 'filter_$', 'year_admission'
], axis=1)
```

```python
df.info()
```

```python
df.drop(df[df['gender'] == "missing"].index, inplace = True)
df.drop(df[df['etnia'] == "missing"].index, inplace = True)
df.drop(df[df['Oxygen_saturation_reduced'] == "missing"].index, inplace = True)
df.drop(df[df['death'] == "missing"].index, inplace = True)
df.drop(df[df['codigoRecebeuVacina'] == "unavailable"].index, inplace = True)
df.drop(df[df['codigoRecebeuVacina'] == "ignorado"].index, inplace = True)
df.drop(df[df['Oxygen_saturation_reduced'] == "ignorado"].index, inplace = True)
df['gender'] = df['gender'].map({'male':0,'female':1})
df['comorbidades_cat'] = df['comorbidades_cat'].map({'nao':0,'uma':1,'duas':2,'tres ou mais':3})
df['death'] = df['death'].map({'nao':0,'sim':1})
df['codigoRecebeuVacina'] = df['codigoRecebeuVacina'].map({'nao':0,'sim':1})
df.drop(df[df['Doses_vacina'] == "missing"].index, inplace = True)
df['Doses_vacina'] = df['Doses_vacina'].map({'zero':0,'uma':1,'duas':2,'tres':3})
df['comorbidades_dic'] = df['comorbidades_dic'].map({'nao':0,'sim':1})
df['Oxygen_saturation_reduced'] = df['Oxygen_saturation_reduced'].map({'nao':0,'sim':1})
df['fever'] = df['fever'].map({'nao':0,'sim':1})
df['cough'] = df['cough'].map({'nao':0,'sim':1})
df['dyspnea'] = df['dyspnea'].map({'nao':0,'sim':1})
```

```python
df['idade_cat_3'] = df['idade_cat_3'].map({'<2':0,'2-11':1, '12-17':2})
df['idade_cat_4'] = df['idade_cat_4'].map({'<2':0,'2-4':1, '5-11':2, '12-17':3})
df['etnia'] = df['etnia'].map({'Brown':0, 'White':1, 'Indigenous':2, 'Black':3, 'Asian':4})
df['region'] = df['region'].map({'Northeast':0, 'South':1, 'Southeast':2, 'North':3, 'Central_West':4})
df['Virus_strain'] = df['Virus_strain'].map({'gamma':0, 'delta':1, 'omicron':2})
df['nosocomial'] = df['nosocomial'].map({'nao':0, 'sim':1})
```

```python
df = df.replace({'sim': 1, 'nao': 0})
```

```python
df[['gender',
 'idade_cat_3',
 'idade_cat_4',
 'etnia',
 'region',
 'Virus_strain',
 'dyspnea',
 'fever',
 'cough',
 'odynophagia',
 'diarrhea',
 'vomit',
 'abdominal_Pain',
 'ageusia',
 'anosmia',
 'REspDistress',
 'Oxygen_saturation_reduced',
 'diabetes_mellitus',
 'obesity',
 'cardiology',
 'pulmonary',
 'hypertension',
 'immunesupression',
 'renal',
 'asthma',
 'comorbidades_dic',
 'comorbidades_cat',
 'death',
 'codigoRecebeuVacina',
 'Doses_vacina',
 'hematology',
 'neurology',
 'oncology',
 'Transplant',
 'sind_down',
 'syndrome',
 'nosocomial']] = df[['gender',
 'idade_cat_3',
 'idade_cat_4',
 'etnia',
 'region',
 'Virus_strain',
 'dyspnea',
 'fever',
 'cough',
 'odynophagia',
 'diarrhea',
 'vomit',
 'abdominal_Pain',
 'ageusia',
 'anosmia',
 'REspDistress',
 'Oxygen_saturation_reduced',
 'diabetes_mellitus',
 'obesity',
 'cardiology',
 'pulmonary',
 'hypertension',
 'immunesupression',
 'renal',
 'asthma',
 'comorbidades_dic',
 'comorbidades_cat',
 'death',
 'codigoRecebeuVacina',
 'Doses_vacina',
 'hematology',
 'neurology',
 'oncology',
 'Transplant',
 'sind_down',
 'syndrome',
 'nosocomial']].astype(int)
```

```python
df = df.drop(['idade_cat_3', 'idade_cat_4', 'comorbidades_dic', 'comorbidades_cat'], axis=1)
```

```python
from imblearn.combine import SMOTEENN
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
y=df['death']
X=df.drop(['death'],axis=1)

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
smtenn = SMOTEENN()
smt = SMOTE()

# from sklearn.preprocessing import StandardScaler

# #scaler = StandardScaler()
# #X_train2 = scaler.fit_transform(X_train2)
# #X_test2 = scaler.transform(X_test2)

X_train_smote, y_train_smote = smtenn.fit_resample(X_train,y_train)

from sklearn.linear_model import LogisticRegression
from sklearn import metrics
logit_model = LogisticRegression()
logit_model.fit(X_train_smote,y_train_smote)

from sklearn.model_selection import cross_val_score
resultado_cv = cross_val_score(logit_model, X_train_smote, y_train_smote, cv=10)

resultado_cv
```

```python
prediction=logit_model.predict(X_test)
print('The accuracy of the Logistic Regression is',metrics.accuracy_score(prediction,y_test))
```

```python
coefficients = logit_model.coef_[0]

feature_importance = pd.DataFrame({'Feature': X.columns, 'Importance': np.abs(coefficients)})
feature_importance = feature_importance.sort_values('Importance', ascending=True)
feature_importance.plot(x='Feature', y='Importance', kind='barh', figsize=(15, 15))
```
