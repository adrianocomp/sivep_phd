# seleção para Suporte Ventilatório

```python
import pandas as pd
import numpy as np
df = pd.read_csv("bases_de_dados/Sivep/bases_novas/sivep_kids_up_for_vent.csv")
## Separando o dataset

X=df.iloc[:,:-1]
y=df.iloc[:,-1]
#y=df['outcome_final_SIVEP']

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.feature_selection import chi2
res1 = chi2(X, y)
pvalues = pd.Series(res1[1])
pvalues.index = X_train.columns
pvalues.sort_index()
```

```python
scores, pvalues = chi2(X_train, y_train)
for i in range(len(scores)):
    print(f"Feature {i}: P-value = {pvalues[i]}, Chi2-statistic = {scores[i]}")
```

```python
# xgboost for feature importance on a classification problem
#com dados completos
from xgboost import XGBClassifier
from matplotlib import pyplot as plt
# define dataset

# define the model
model = XGBClassifier()
# fit the model
model.fit(X, y)
# get importance
importance = model.feature_importances_
# summarize feature importance
#for i,v in enumerate(importance):
#	print('Feature: %0d, Score: %.5f' % (i,v))
# plot feature importance
#pyplot.bar([x for x in range(len(importance))], importance)
#pyplot.show()

feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nsmallest(39).plot(figsize=(10,10), kind='barh', fontsize=10)
plt.xlabel("Importance Scores", fontsize=10)
plt.ylabel("Carachteristic", fontsize=10)
plt.show()
```

```python
#com o dataset inteiro

from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=500, n_jobs=-1, random_state=42)
model.fit(X, y)
# plotting feature importances
features = df.drop('suport_vent', axis=1).columns
importances = model.feature_importances_
indices = np.argsort(importances)
plt.figure(figsize=(10,10))
#plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices], fontsize=14)
plt.xlabel('MeanDeacreaseGini', fontsize=14)
plt.ylabel('Characteristics',fontsize=14)
plt.show()
```
