```python
scores, pvalues = chi2(X_train, y_train)
```

```python
for i in range(len(scores)):
    print(f"Feature {i}: P-value = {pvalues[i]}, Chi2-statistic = {scores[i]}")
```

```python
from sklearn.feature_selection import SelectKBest
import matplotlib.pyplot as plt
bestfeatureschi2 = SelectKBest(score_func=chi2)
fit = bestfeatureschi2.fit(X_train,y_train)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X_train.columns)
#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Caracteristicas','Pontuação']  #naming the dataframe columns
print(featureScores.nlargest(34,'Pontuação'))  #print N best features
#plot
feat_importances = pd.Series(fit.scores_, index=X_train.columns)
feat_importances.nsmallest(34).plot(figsize=(10,10), kind='barh', fontsize=12)
plt.xlabel('Pontuação', fontsize=12)
plt.ylabel('Características', fontsize=12)
plt.show()
```

# Feature selection com XGBoost

```python
# xgboost for feature importance on a classification problem
#com dados de treinamento
from xgboost import XGBClassifier
from matplotlib import pyplot
# define dataset

# define the model
model = XGBClassifier()
# fit the model
model.fit(X_train, y_train)
# get importance
importance = model.feature_importances_
# summarize feature importance
#for i,v in enumerate(importance):
#	print('Feature: %0d, Score: %.5f' % (i,v))
# plot feature importance
#pyplot.bar([x for x in range(len(importance))], importance)
#pyplot.show()

feat_importances = pd.Series(model.feature_importances_, index=X_train.columns)
feat_importances.nsmallest(39).plot(figsize=(10,10), kind='barh', fontsize=14)
plt.xlabel("Importance Scores", fontsize=14)
plt.ylabel("Carachteristic", fontsize=14)
plt.show()
```

```python
# xgboost for feature importance on a classification problem
#com dados completos
from xgboost import XGBClassifier
from matplotlib import pyplot
# define dataset

# define the model
model = XGBClassifier()
# fit the model
model.fit(X, y)
# get importance
importance = model.feature_importances_
# summarize feature importance
#for i,v in enumerate(importance):
#	print('Feature: %0d, Score: %.5f' % (i,v))
# plot feature importance
#pyplot.bar([x for x in range(len(importance))], importance)
#pyplot.show()

feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nsmallest(39).plot(figsize=(10,10), kind='barh', fontsize=14)
plt.xlabel("Importance Scores", fontsize=14)
plt.ylabel("Carachteristic", fontsize=14)
plt.show()
```

# Feature Selection with Random Forest

```python
#com train test split

from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=500, n_jobs=-1, random_state=42)
model.fit(X_train, y_train)
# plotting feature importances
features = df.drop('outcome_final_SIVEP', axis=1).columns
importances = model.feature_importances_
indices = np.argsort(importances)
plt.figure(figsize=(10,10))
#plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices], fontsize=14)
plt.xlabel('MeanDeacreaseGini', fontsize=14)
plt.ylabel('Characteristics',fontsize=14)
plt.show()
```

```python
#com o dataset inteiro

from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=500, n_jobs=-1, random_state=42)
model.fit(X, y)
# plotting feature importances
features = df.drop('outcome_final_SIVEP', axis=1).columns
importances = model.feature_importances_
indices = np.argsort(importances)
plt.figure(figsize=(10,10))
#plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices], fontsize=14)
plt.xlabel('MeanDeacreaseGini', fontsize=14)
plt.ylabel('Characteristics',fontsize=14)
plt.show()
```
